## 3.3 Списочные выражения. Модель памяти для типов

A. Список квадратов

**Формат ввода**

Даны две переменные a и b в отдельных строках со знаком равенства.


**Формат вывода**

Напишите одно списочное выражение, которое формирует список квадратов всех целых чисел от [a,b] включительно.

```python
a = 1
b = 5
[x ** 2 for x in range(a, b + 1)]
```

B - Список квадратов 2

**Формат ввода**

САписок квадратов должен строиться вне зависимости от порядка значений a и b.

Если a < b, счёт идёт вперёд. Если a > b, счёт идёт назад.

**Формат вывода**

Один список, содержащий квадраты всех чисел от a до b включительно, в соответствующем порядке.

```python
a = 5
b = -4

[x ** 2 for x in range(a, b + (1 if a < b else -1), (1 if a < b else -1))]
```

C - Основы фильтрации

**Формат ввода**

Цель — выбрать из диапазона только те числа, которые делятся на заданное значение без остатка.
Вашему решению даны три переменные: a, b и d.
Сформируйте список всех чисел, кратных d, лежащих в диапазоне от a до b включительно.

**Формат вывода**

С помощью одного списочного выражения постройте список всех чисел от a до b, которые делятся на d без остатка.

```python
a = -9
b = 15
d = 3

[x  for x in range(a, b + 1) if x % d == 0]
```

D - Множество нечетных чисел

**Формат ввода**

Вашему решению предоставлен список numbers, содержащий натуральные числа.
Нужно выбрать из него только нечётные числа — и сохранить их в множестве, чтобы исключить повторы.

**Формат вывода**

Сформируйте одно выражение, которое создаёт множество всех нечётных чисел из списка numbers.

```python
numbers = [1, 2, 3, 2, 1, 2, 3, 2, 1, 2]
set(x for x in numbers if x % 2 == 1)
```

E - Множество всех полных квадратов

**Формат ввода**

Полный квадрат — это натуральное число, являющееся квадратом другого натурального числа.

Например:

4 — это 2 ** 2,
49 — это 7 ** 2.
Ваша задача — из заданного списка numbers отобрать только полные квадраты и собрать их в множество (без повторов).

Вашему решению будет предоставлен список numbers, содержащий натуральные числа.

**Формат вывода**

Сформируйте одно выражение, которое создаёт множество всех полных квадратов из списка.

```python
numbers = [number for number in range(16, 100, 4)]
{x for x in numbers if (x ** 0.5).is_integer() is True}
```

F - Длины всех слов

**Формат ввода**

В этой задаче нужно посчитать длину каждого слова в строке.

Вашему решению будет предоставлена строка sentence, в которой слова разделены пробелами.

**Формат вывода**

Напишите списочное выражение, которое создаёт список длин этих слов.

```python
sentence = 'Ехали медведи на велосипеде'
[len(x) for x in sentence.split()]
```

G - Цифровая выжимка

**Формат ввода**

Вашему решению будет предоставлена переменная text — строка, которая может содержать любые символы.

**Формат вывода**

Нужно извлечь все цифры и собрать их в одну строку, сохранив порядок появления.

```python
text = '33 коровы,\n' + \
    '33 коровы,\n' + \
    '33 коровы -\n' + \
    'Свежая строка.\n' + \
    '33 коровы,\n' + \
    'Стих родился новый,\n' + \
    'Как стакан парного молока.\n' + \
    'Стих родился новый,\n' + \
    'Как стакан парного молока.\n'
''.join(char for char in text if char.isdigit() is True)
```

H - Аббревиатура

**Формат ввода**

Вам дана строка string, содержащая слова, разделённые пробелами.

**Формат вывода**

Нужно составить строку из первых букв всех слов, в верхнем регистре, без пробелов.

```python
string = 'открытое акционерное общество'
''.join(x[0].upper() for x in string.split())
```

I - Преобразование в строку

**Формат ввода**

Допустим, у вас есть список натуральных чисел, возможно с повторениями.

Ваша задача — превратить его в строку, в которой будут только уникальные числа, отсортированные по возрастанию, разделённые через дефис с пробелами (' - ').

Вашему решению предоставлен список натуральных чисел numbers.

**Формат вывода**

Сформируйте строку по описанным правилам.

```python
numbers = [1, 1, 3, 1, 10, 2, 4, 6, 7, 1, 2, 7]
' - '.join(str(x) for x in sorted(set(numbers)))
```

J = Огласите список

**Формат ввода**

Вашему решению предоставлена строка words.
В русском языке гласными являются: аяуюоёэеиы
В английском: aeiouy

**Формат вывода**

Напишите выражение, которое создаёт список слов, в которых не менее трёх гласных букв (в любом регистре).

```python
words = 'My homework is too difficult!'
[x for x in words.split() if sum(1 for char in x if char.lower() in set('aeiouyаяуюоёэеиы')) >= 3]
```

K = Выявление уникальности

**Формат ввода**

Вашему решению предоставлен список numbers.

**Формат вывода**

Напишите выражение, которое создаёт множество всех чисел, встречающихся в списке ровно один раз.

```python
numbers = [-8, 7, -3, -2, 5, 0, 7, -2, -8, 6, -10, 4, -9, -9, 7]
{x for x in numbers if sum(1 for y in numbers if x == y) == 1}
```

L - Максимальное произведение

**Формат ввода**

Вашему решению предоставлено множество numbers.

**Формат вывода**

Напишите одно выражение, которое находит максимальное произведение двух различных чисел из этого множества.

```python
numbers = {2, 4, 5, 7, -10, -8, 10, -9, -1}
max(first * second for first in numbers for second in numbers if first != second)
```
