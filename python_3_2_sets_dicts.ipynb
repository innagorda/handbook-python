A. Символическая выжимка

**Формат ввода**

Вводится одна строка.

**Формат вывода**

Требуется вывести все символы этой строки без повторений.

```python
input_str = input()
s = set(input_str)
output_str = s
print(''.join(output_str))
```

C - Зайка - 8

**Формат ввода**

В первой строке записано натуральное число N — количество выделенных придорожных местностей.
В каждой из N последующих строк записано описание придорожной местности.

**Формат вывода**

Вывести все найденные объекты в придорожных местностях.

```python
row_cont = int(input())
objects = set()

for i in range(0, row_cont):
    obj_set = input().split(' ')
    objects = objects | set(obj_set)

print('\n'.join(objects))
```

D - Кашееды

**Формат ввода**

В первых двух строках указывается количество детей, любящих манную и овсяную каши ( N и 
M).
Затем идут N строк — фамилии детей, которые любят манную кашу, и M строк с фамилиями детей, любящих овсяную кашу.
Гарантируется, что в группе нет однофамильцев.

**Формат вывода**

Количество учеников, которые любят обе каши.
Если таких не окажется, в строке вывода нужно написать «Таких нет».

```python
set1_cnt = int(input())
set2_cnt = int(input())

objects, set1, set2 = set(), set(), set()

for i in range(0, set1_cnt):
    set1.add(input())
for i in range(0, set2_cnt):
    set2.add(input())
    
objects = set1 & set2

if len(objects) > 0:
    print(len(objects))
else:
    print('Таких нет')
```

E - Кашееды 2

**Формат ввода**

В первых двух строках указывается количество детей, любящих манную и овсяную каши ( N и 
M).
Затем идут N+M строк — перемешанные фамилии детей.
Гарантируется, что в группе нет однофамильцев.

**Формат вывода**

Количество учеников, которые любят только одну кашу.
Если таких не окажется, в строке вывода нужно написать «Таких нет».

```python
set1_cnt = int(input())
set2_cnt = int(input())

students = dict()
porrige1 = []

for i in range(0, set1_cnt + set2_cnt):
    surname = input()
    if surname in students:
        students[surname] += 1
    else:
        students[surname] = 1
    
porrige1 = porrige1 + [key for key, value in students.items() if value == 1]
    
if len(porrige1) > 0:
    print(len(porrige1))
else:
    print('Таких нет')
```

F - Кашееды 3

**Формат ввода**

В первых двух строках указывается количество детей, любящих манную и овсяную каши ( N и 
M).
Затем идут N+M строк — перемешанные фамилии детей.
Гарантируется, что в группе нет однофамильцев.

**Формат вывода**

В алфавитном порядке фамилии учеников, которые любят только одну кашу.

```python
set1_cnt = int(input())
set2_cnt = int(input())

students = dict()
porrige1 = []

for i in range(0, set1_cnt + set2_cnt):
    surname = input()
    if surname in students:
        students[surname] += 1
    else:
        students[surname] = 1
    
porrige1 = porrige1 + [key for key, value in students.items() if value == 1]
if len(porrige1) > 0:
    porrige1.sort()
    print('\n'.join(porrige1))
else:
    print('Таких нет')

```

H - Кашееды - 4

**Формат ввода**

В первой строке задаётся количество детей в группе (N).
В следующих N строках записана фамилия ребенка и список его любимых каш.
В последней строке записана каша, информацию о которой хочет получить воспитатель.

**Формат вывода**

Фамилии учеников, которые любят заданную кашу, в алфавитном порядке.
Если таких не окажется, в строке вывода нужно написать «Таких нет».

```python
lines_cnt = int(input())
porrige = dict()
names_list = []

for i in range(0, lines_cnt):
    name, *food = input().split()
    for por in food:
        if por in porrige:
            porrige[por].append(name)
        else:
            porrige[por] = [name]

search_text = input()

for key, value in porrige.items():
    if key == search_text:
        names_list = value

if len(names_list) > 0:
    names_list.sort()
    for v in names_list:
        print(v)
else:
    print('Таких нет')

```
