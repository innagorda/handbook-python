### J - Частотный анализ на минималках

Напишите программу, которая читает строки до тех пор, пока не встретится строка 'ФИНИШ', и выводит наиболее часто встречающуюся букву (в нижнем регистре).

**Формат ввода**

Вводятся строки, пока не будет введена строка «ФИНИШ».

```
Баобаб
Белка
Бобы
ФИНИШ
```

**Формат вывода**

Выводится один символ в нижнем регистре — наиболее часто встречающийся во введённой строке.

```
б
```

````python
letters_dist = []
letters_count = []
while (input_str := input()) != 'ФИНИШ':
    letters = list(input_str.lower())
    for item in letters:
        if (item.isalpha() is True):
            try:
                index = letters_dist.index(item) 
                letters_count[index] += 1
            except ValueError:
                letters_dist.append(item)
                letters_count.append(1)

indices = [i for i, x in enumerate(letters_count) if x == max(letters_count)]
freq_letters = [x for i, x in enumerate(letters_dist) if i in indices]
freq_letters.sort()
print(freq_letters[0])
````

**Примечание**
Самый распространенный способ поиска НОД — Алгоритм Эвклида.

````python
n_list = input().split()
n_list = [int(x) for x in n_list]
````

````python
max_n = max(n_list[0], n_list[1])
min_n = min(n_list[0], n_list[1])
nod = n_list[0]

for index in range(1, len(n_list)):
    
    max_n = max(nod, n_list[index])
    min_n = min(nod, n_list[index])

    while min_n != 0: 
        max_n, min_n = min_n, max_n % min_n
        nod = max_n # последний остаток сохраняем


print(nod)
````

````python
# Вариант с рекурсией

def calc_gcd(a, b):
    """ Алгоритм Евклида
GCD равен a, если a и b совпадают.
GCD равен GCD от a - b и b, если a больше b.
GCD равен GCD от a и b - a, если a меньше b."""
    count_of_execution = 0

    def f_inner(a, b, count_of_execution):
        count_of_execution += 1
        min_n, max_n = min(a, b), max(a, b)
        if min_n == max_n:
            return max_n, count_of_execution
        else:
            return f_inner(max_n-min_n, min_n, count_of_execution)

    return f_inner(a, b, count_of_execution)
````

### T - Польский калькулятор — 2

Операции, которые выполняются с одним значением, называются унарными, с двумя — бинарными, с тремя — тернарными.

Давайте улучшим наш калькулятор, добавив поддержку следующих операций:

- бинарные:
 ```   
    + (сложение),
    - (вычитание),
    * (умножение),
    / (деление нацело; для отрицательных чисел работает по тем же правилам, что и в Python);
```
- унарные:
 ``` 
    ~ (унарный минус — меняет знак),
    ! (факториал),
    # (клонирование — вернуть в стек значение дважды);
 ``` 
- тернарные:
 ``` 
@ (возвращает в стек те же три значения, но в ином порядке: второе, третье, первое).
 ``` 

**Формат ввода**

Вводится одна строка, содержащая разделённые пробелами целые числа и знаки операций. Вместе они составляют корректное выражение в обратной польской нотации, не содержащее деления на ноль и взятия факториала от отрицательного числа.

```
7 1 10 100 # * @ - + + ~
```

**Формат вывода**

Выводится одно целое число — результат вычисления выражения.

```
-10016
```
